```python
import pathlib
import textwrap
import os
import google.generativeai as genai
from IPython.display import Markdown, display

def to_markdown(text):
  text = text.replace(".","*")
  return Markdown(textwrap.indent(text, '>', predicate = lambda _: True))

from google.colab import userdata
# It's recommended to securely store and retrieve API keys, e.g., using Colab secrets.
os.environ['GOOGLE_API_KEY']='AIzaSyA12CNIzcPWeA7xqY321jIn5h1RmMkYrlE'
genai.configure(api_key=os.environ['GOOGLE_API_KEY']) # Re-enabling genai.configure
for m in genai.list_models(): # Uncommenting to list models
  if 'generateContent' in m.supported_generation_methods: # Fixed typo
    print(m.name)
```

```python
model = genai.GenerativeModel('gemini-2.5-flash')
chat = model.start_chat(history=[])

print("Chat with the AI model! Type 'exit' to end the conversation.")
while True:
    user_input = input("You: ")
    if user_input.lower() == 'exit':
        break
    response = chat.send_message(user_input)
    display(to_markdown(response.text))
```
